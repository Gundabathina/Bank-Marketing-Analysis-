# Outlier Detection

```python

from collections import Counter

def detect_outliers(data, features):
    outlier_indices = []

    for c in features:
        # 1st quartile
        Q1 = np.percentile(data[c], 25)
        # 3rd quartile
        Q3 = np.percentile(data[c], 75)
        # IQR
        IQR = Q3 - Q1
        # Outlier step
        outlier_step = IQR * 1.5

        # detect outliers and their indices
        outlier_list_col = data[(data[c] < Q1 - outlier_step) | (data[c] > Q3 + outlier_step)].index
        # store indices
        outlier_indices.extend(outlier_list_col)

    # Count the number of outliers per index
    outlier_indices = Counter(outlier_indices)
    # Get indices for multiple outliers
    multiple_outliers = list(i for i, v in outlier_indices.items() if v > 1)

    return multiple_outliers
```

```python

# Assuming 'data' is your DataFrame and 'detect_outliers' function is already defined
outliers = detect_outliers(data, ['age', 'day', 'duration', 'campaign', 'previous'])

# Locate the rows with outliers
outlier_data = data.loc[outliers]

print(outlier_data)
```

# Algorithm Works

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import cohen_kappa_score, classification_report, confusion_matrix, ConfusionMatrixDisplay, f1_score, accuracy_score

# Assuming 'data' is your DataFrame

# Step 1: Identify numeric and categorical columns
numeric_columns = data.select_dtypes(include=[np.number]).columns
categorical_columns = data.select_dtypes(include=[object]).columns

# Step 2: Encode categorical columns
data_encoded = pd.get_dummies(data[categorical_columns])

# Step 3: Combine numeric and encoded categorical columns
data_processed = pd.concat([data[numeric_columns], data_encoded], axis=1)

# Step 4: Split features and target variable
X = data_processed.iloc[:, :-1]  # Assuming the last column is the target
Y = data_processed.iloc[:, -1]

# Step 5: Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Step 6: Standardize the numeric features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Step 7: Initialize and fit the classifier
classifier = RandomForestClassifier(random_state=42)
classifier.fit(X_train, y_train)

# Step 8: Make predictions
y_pred = classifier.predict(X_test)

# Step 9: Calculate metrics
accuracies = accuracy_score(y_test, y_pred)
kappaScores = cohen_kappa_score(y_test, y_pred)
f1scores = f1_score(y_test, y_pred, average='weighted')

# Step 10: Print metrics
print("Accuracy:", accuracies)
print("Cohen's Kappa:", kappaScores)
print("F1 Score:", f1scores)

# Classification report
print(classification_report(y_test, y_pred))

# Confusion matrix
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
plt.show()
```
